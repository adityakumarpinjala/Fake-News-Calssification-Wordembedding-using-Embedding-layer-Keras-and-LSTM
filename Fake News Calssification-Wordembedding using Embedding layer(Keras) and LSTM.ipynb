```python
import pandas as pd
import numpy as np
import nltk
from nltk.stem.porter import PorterStemmer
import re
from nltk.corpus import stopwords
```

```python
X_tdata=pd.read_csv("train.csv")
```

```python
X_tdata.head()
```

```python
X_tdata.isnull().sum()
```

```python
X_tdata.shape
```

```python
X_tdata.dropna(subset=['title'],inplace=True)
```

```python
X_tdata.shape
```

```python
X_tdata1=X_tdata
```

```python
X_tdata.head()
```

```python
ps=PorterStemmer()
```

```python
def preprocesing(snts):
    Docs=snts.lower()
    Docs=re.sub("[^A-Za-z]", " ",Docs)
    Docs=Docs.split()
    words=[ps.stem(word) for word in Docs if word not in stopwords.words("english")]
    Docs=" ".join(words)
    return Docs
```

```python
X_tdata1['title']=X_tdata1['title'].apply(lambda x : preprocesing(x))
```

```python
X_tdata1.tail()
```

```python
X_tdata1.reset_index(inplace=True,drop=True)
```

```python
X_tdata1.tail()
```

```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import Dense
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
```

```python
messages=X_tdata1.copy()
```

```python
messages
```

```python
corpus=list(messages['title'])
```

```python
corpus[:4]
```

```python
vocab_size=5000
#index from the dictionary
one_repre=[one_hot(snt,vocab_size) for snt in corpus]
one_repre
```

```python
sent_length=20
embedded_docs=pad_sequences(one_repre,padding='pre',maxlen=sent_length)
```

print(embedded_docs)

```python
len(embedded_docs)
```

```python
#Creating the model
embedding_vector_features=40
model=Sequential()
model.add(Embedding(vocab_size,embedding_vector_features,input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
len(embedded_docs)
```

```python
X_final=np.array(embedded_docs)
y=messages['label']
Y_final=np.array(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X_final,Y_final,random_state=200,test_size=0.25)
```

```python
model.fit(X_train,Y_train,validation_data=(X_test,Y_test),epochs=10,batch_size=64)
```

```python
Y_pred=model.predict_classes(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm=confusion_matrix(Y_pred,Y_test)
```
